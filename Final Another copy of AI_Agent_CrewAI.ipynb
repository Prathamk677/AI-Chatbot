```python
import os
import google.generativeai as genai

# Set your Gemini API Key securely in the environment
os.environ["GOOGLE_API_KEY"] = "AIzaSyA9p5nKZS_S5T-pVL9O1fi2JJlg1diC3FY"
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

# Initialize the Gemini model (Flash)
model = genai.GenerativeModel("gemini-2.0-flash-lite-001")

# Define the two agents
class Agent:
    def __init__(self, role, goal, backstory):
        self.role = role
        self.goal = goal
        self.backstory = backstory

    def act(self, prompt):
        """Simulate the agent acting based on the prompt."""
        print(f"\n{self.role} is performing the task: {self.goal}\n")
        response = model.generate_content(prompt)
        return response.text


# Agent 1: Professor (to generate the MCQs)
professor = Agent(
    role="Professor",
    goal="Create 5 multiple choice questions (MCQs) on Artificial Intelligence.",
    backstory="You are an experienced professor who specializes in Artificial Intelligence and is tasked with creating MCQs for students."
)

# Agent 2: AI Expert (to review and enhance the MCQs)
ai_expert = Agent(
    role="AI Expert",
    goal="Review and enhance the generated MCQs, make them harder, and provide correct answers with explanations.",
    backstory="You are an AI expert who specializes in improving educational content, making it more challenging and insightful."
)

# Define the tasks for each agent
def generate_mcqs():
    """Generate 5 multiple-choice questions (MCQs) about Artificial Intelligence."""
    prompt1 = (
        "You are a professor. Create 5 multiple choice questions (MCQs) on Artificial Intelligence. "
        "Each question should have 4 options and only one correct answer. Do not include answers yet."
    )
    return professor.act(prompt1)

def review_mcqs(mcqs):
    """Review and enhance the generated MCQs, making them harder and adding answers with explanations."""
    prompt2 = (
        f"You are an AI expert. Review the following 5 MCQs, make them harder if necessary, "
        "and provide the correct answers with brief explanations for each.\n\n"
        f"{mcqs}"
    )
    return ai_expert.act(prompt2)

# Generate the initial MCQs using Professor
print("ğŸ“ Generating MCQs...\n")
mcqs = generate_mcqs()

# If MCQs were successfully generated, pass them to the AI Expert for review
if mcqs:
    print("MCQs Generated:\n")
    print(mcqs)

    # Review and enhance the MCQs
    print("\nğŸ” Reviewing MCQs and adding answers...\n")
    final_output = review_mcqs(mcqs)

    if final_output:
        print("\nFinal MCQs with Answers and Explanations:\n")
        print(final_output)
else:
    print("No MCQs were generated. Please check for errors.")
```
